```python
import pandas as pd
import matplotlib.pyplot as plt

USERS_FILE = "users_data_enriched.csv"

# Load users dataset
users = pd.read_csv(USERS_FILE)

print("Users shape:", users.shape)
print(users.head())


if "goal" in users.columns:
    goal_counts = users["goal"].value_counts()

    plt.figure()
    plt.bar(goal_counts.index.astype(str), goal_counts.values)
    plt.title("Users Goals Distribution")
    plt.xlabel("Goal")
    plt.ylabel("Number of Users")
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
else:
    print("No 'goal' column found in users_data_enriched.csv")


if "age" in users.columns:
    ages = pd.to_numeric(users["age"], errors="coerce").dropna()

    plt.figure()
    plt.hist(ages, bins=10)
    plt.title("User Ages Distribution")
    plt.xlabel("Age")
    plt.ylabel("Number of Users")
    plt.tight_layout()
    plt.show()
else:
    print("No 'age' column found in users_data_enriched.csv")
```
